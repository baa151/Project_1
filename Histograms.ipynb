```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from CSV
cvsfileTX = 'TX_death2021.csv'
dataTX = np.genfromtxt(cvsfileTX, delimiter=',')

# Extract data columns
fips_TX = dataTX[:, 0]
death_TX = dataTX[:, 1]

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(fips_TX, death_TX, c='red', marker='o', alpha=0.7, label='Deaths in TX')

# Customize the plot
plt.title('COVID-19 Deaths in Texas (2021)')
plt.xlabel('FIPS Code')
plt.ylabel('Number of Deaths')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the scatter plot
plt.show()

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(death_TX, bins=25, color='blue', alpha=0.7, label='Deaths in TX')
plt.title('Histogram of COVID-19 Deaths in Texas (2021)')
plt.xlabel('Number of Deaths')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the histogram
plt.show()

# Plot density plot (KDE)
plt.figure(figsize=(10, 6))
sns.histplot(death_TX, kde=True, color='green', fill=True)
plt.title('Density Plot of COVID-19 Deaths in Texas (2021)')
plt.xlabel('Number of Deaths')
plt.ylabel ('Density')
plt.grid(True, linestyle='--', alpha=0.6)

# Show the density plot
plt.show()
```
