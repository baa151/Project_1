# QUESTION ONE

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from CSV
cvsfile3 = 'TX_death2021.csv'
data3 = np.genfromtxt(cvsfile3, delimiter=',')

# Extract data columns
fips_TX = data3[:, 0]
death_TX = data3[:, 4]

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(fips_TX, death_TX, c='red', marker='o', alpha=0.7, label='Deaths in TX')

# Customize the plot
plt.title('COVID-19 Deaths in Texas (2021)')
plt.xlabel('FIPS Code')
plt.ylabel('Number of Deaths')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from CSV
cvsfile6 = 'TX_GDP2021.csv'
data6 = np.genfromtxt(cvsfile6, delimiter=',')

# Extract data columns
fips_CA6 = data6[:, 0]
GDP_CA6 = data6[:, 1]

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(fips_CA6, GDP_CA6, c='green', marker='o', alpha=0.7, label='GDP in TX')

# Customize the plot
plt.title('Gross Domestic Product (GDP) in Texas (2021)')
plt.xlabel('FIPS Code')
plt.ylabel('GDP (Millions of Dollars)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```
